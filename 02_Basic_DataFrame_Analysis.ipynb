```python
state_population_dataframe = pd.read_csv("1.section data/nst-est2020.csv")
print(state_population_dataframe)
```

```python
state_old_column_names = state_population_dataframe.columns
state_new_column_names = [name.lower() for name in state_old_column_names]
print(state_new_column_names)
```

```python
column_names_dict = {}
index_position = 0
for name in state_old_column_names:
    column_names_dict[name] = state_new_column_names[index_position]
    index_position += 1
    
print(column_names_dict)
```

```python
state_population_dataframe.rename(columns=column_names_dict, inplace=True)
print(state_population_dataframe)
```

```python
state_population = state_population_dataframe.tail(n=52)
print(state_population)
```

```python
print(state_population.sum(numeric_only=True))
```
